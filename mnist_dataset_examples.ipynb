```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import layers
```

```python
(x_train, y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train/255
x_test = x_test/255
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
plt.imshow(x_train[10000], cmap='Greys')
```

```python
model_rahul = tf.keras.models.Sequential([
    layers.Input(x_train.shape[1:]),
    layers.Flatten(),
    layers.Dense(10,activation='softmax')
])
model_rahul.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
model_rahul.summary()
```

```python
y_onehot_train = tf.one_hot(y_train,10)
model_rahul.fit(x_train,y_onehot_train)
```

```python
model_rahul = tf.keras.models.Sequential([
    layers.Input(x_train.shape[1:]),
    layers.Flatten(),
    layers.Dense(10,activation='softmax')
])
model_rahul.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_rahul.summary()
```

```python
histroy_rahul= model_rahul.fit(x_train,y_train, epochs=20, batch_size=128, validation_data=(x_test,y_test))
```

```python
histroy_rahul.history
```

```python
plt.plot(histroy_rahul.history['loss'],label='train')
plt.plot(histroy_rahul.history['val_loss'],label='val')
plt.ylabel('loss')
plt.legend()
plt.show()

plt.plot(histroy_rahul.history['accuracy'],label='train')
plt.plot(histroy_rahul.history['val_accuracy'],label='val')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
model_rahul.evaluate(x_test,y_test)
```

```python
probs=model_rahul.predict(x_test[:5])
preds=np.argmax(probs, axis=1)
for i in range(5):
  print(probs[i], "=>", preds[i])
  plt.imshow(x_test[i],cmap="Greys")
  plt.show()
```
